```python
#merging data
data_1 = pd.DataFrame({
    "name":["Jayesh", "Dinesh", "jasmine"],
    "roll":[1,2,3]
})

data_2 = pd.DataFrame({
    "name":["Ritesh", "Arpita", "Aaditya"],
    "roll":[4,5,6]
})
print("\nfirst dataset : \n", data_1, "\n second dataset : \n", data_2)

#concat
pd.concat([data_1, data_2])
#returns a dataset which have data of both datasets
#Union


#Merge 
# merges the contents of both datasets but only includes those contents are in both datasets
#Instersection
pd.merge(data_1, data_2, on="roll")

```
