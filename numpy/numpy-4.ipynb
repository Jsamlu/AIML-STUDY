```python
import numpy as np
import matplotlib.pyplot as plt
import streamlit as st
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.random.rand(3,3)
arr3 = np.zeros((4,4))

np.save("array1.npy", arr1)
np.save("array2.npy", arr2)
np.save("array3.npy", arr3) # creates a bin/ array file 
```

```python
#Load the saved Array

Loaded_array1 = np.load("array1.npy")
Loaded_array2 = np.load("array2.npy")
Loaded_array3 = np.load("array3.npy")
print(Loaded_array1)
print(Loaded_array2)
print(Loaded_array3)
```

```python
#Load a file and display and load a dark-version

try:
    logo = np.load("numpy_logo.npy")
    #Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("NumPy")
    plt.grid(False)

except FileNotFoundError:
    print("numpy logo file not found")
```

```python
#Make Dark mode

try:
    logo = np.load("numpy_logo.npy")
    #Display
    plt.figure(figsize=(10, 5))
    plt.subplot(122)
    plt.imshow(1 - logo)
    plt.title("NumPy")
    plt.grid(False)

except FileNotFoundError:
    print("numpy logo file not found")
```
